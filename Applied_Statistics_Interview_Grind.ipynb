###31. What is the covariance of a joint probability distribution?

The covariance of a joint probability distribution measures how two random variables change together.
For two random variables \( X \) and \( Y \), the covariance is defined as:

$$
\text{Cov}(X, Y) = E\left[(X - E[X])(Y - E[Y])\right]
$$

Or equivalently,

$$
\text{Cov}(X, Y) = E[XY] - E[X]E[Y]
$$

* If Cov(X, Y) > 0, the variables tend to increase together.

* If Cov(X, Y) < 0, one variable tends to increase when the other decreases.

* If Cov(X, Y) = 0, they are uncorrelated, but not necessarily independent.
